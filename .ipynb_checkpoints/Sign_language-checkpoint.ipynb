### split the Train, Val, Test sets

```python
df_test = pd.read_csv('/Users/david/Desktop/machine learning/Python Machine Learning books/Python Artificial Intelligence Projects for Beginners/sign_mnist/sign_mnist_test.csv')
```

```python
Xtest = df_test.drop('label', axis=1).values.reshape(7172, 28, 28, 1)
ytest = to_categorical(df_test['label'].values)
```

```python
Xtrain = df.drop('label', axis=1).values.reshape(df.shape[0], 28,28, 1)
ytrain = to_categorical(df['label'].values)
```

```python
Xval, Xtest, yval, ytest = train_test_split(Xtest, ytest, test_size=0.5)
```

```python
print(Xtrain.shape)
print(Xval.shape)
print(Xval.shape)
```

```python
np.unique(np.argmax(yval, axis=1))
```

### build the CNN model

```python
train_gen = ImageDataGenerator(horizontal_flip=True, rescale=1.0 / 255)
train_data = train_gen.flow(Xtrain, ytrain, batch_size=32,shuffle=True)

val_gen = ImageDataGenerator(horizontal_flip=True, rescale=1.0 / 255)
val_data = val_gen.flow(Xval, yval, batch_size=32, shuffle=False)

test_gen = ImageDataGenerator(horizontal_flip=True, rescale=1.0 / 255)
test_data = test_gen.flow(Xtest, ytest, batch_size=32, shuffle=False)
```

```python
model = Sequential()
model.add(Conv2D(filters = 64, kernel_size = 3,  padding='same', activation = 'relu', input_shape = (28,28,1)))
model.add(MaxPooling2D(pool_size = 2))
model.add(Conv2D(filters = 128, kernel_size = 3,  padding='same', activation = 'relu'))
model.add(MaxPooling2D(pool_size = 2))
model.add(Conv2D(filters = 256, kernel_size = 3,  padding='same', activation = 'relu'))
model.add(MaxPooling2D(pool_size = 2))
model.add(Flatten())
model.add(Dense(units = 512, activation = 'relu'))
model.add(Dense(25,activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
epochs = 20
steps_per_epoch = train_data.n//32
validation_steps = val_data.n//32

checkpoint = ModelCheckpoint('model_weights.h5', monitor='val_accuracy',save_weight_only=True, mode='max', verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=0.00001, mode='auto')
callbacks=[PlotLossesKeras(), checkpoint, reduce_lr]

history = model.fit(x=train_data, 
                    steps_per_epoch = steps_per_epoch,
                    epochs=epochs,
                    validation_data=val_data, 
                    validation_steps=validation_steps,
                    callbacks=callbacks)
```

### evaluate the model

```python
model.evaluate(test_data)
```

```python
pred = model.predict(test_data, test_data.n//test_data.batch_size)
```

```python
prediction = np.argmax(pred, axis=1)
```

```python
print(classification_report(np.argmax(ytest, axis=1), prediction))
```

```python
letter = [i[1] for i in letter_list.items()]
```

```python
sns.heatmap(confusion_matrix(np.argmax(ytest, axis=1), prediction), cmap='viridis', annot=True, cbar=False, xticklabels=letter, yticklabels=letter)
fig = plt.gcf()
fig.set_size_inches(20,20)
```
